**Removing Columns that have more than 60% of missing values**

```python
df = df.drop(columns=['Boxoffice','Metacritic Score','Production House','Awards Received'],axis=1)
```

```python
freqgrph=df.select_dtypes(include=['float','int'])
freqgrph.hist(figsize=(20,15))
plt.show()
```

```python
#Some of the columns are skewed some are not. We are using median to replacing the missing values here.
for col in freqgrph.columns:
       df[col]=df[col].fillna(df[col].median())
```

```python
for col in df.select_dtypes(include=['object']):
       df[col]=df[col].fillna(df[col].mode()[0])
```

```python
large_unique_col=['Tags','Country Availability','Director','Writer','Actors','Summary','IMDb Votes']
```

```python
from sklearn.impute import SimpleImputer

# Replace rare categories with the most frequent category
for col in large_unique_col:
   imputer = SimpleImputer(strategy='most_frequent')
   df[col] = imputer.fit_transform(df[[col]]).ravel() #The ravel() function flattens the array into a 1D format.
```

In large datasets where categorical variables have manyÂ uniqueÂ values Frequency Encoding is done by replacing categories with their frequency or count in the dataset.

```python
# Frequency encoding
categorical_col=['Genre','Languages','Country Availability','Director','Writer','View Rating']
for col in categorical_col:
   frequency_encoding = data[col].value_counts()
   data[col] = data[col].map(frequency_encoding)
```

# **Scaling**

```python
#splitting dataset into features and targets
x=data.drop(columns=['Hidden Gem Score'])
y=data['Hidden Gem Score']
```

```python
#applying min mxa scaling on numerical cols
x1=x.copy()
```

```python
#dropping the encoded cols from x1
x1.drop(['Genre', 'Languages', 'Series or Movie', 'Country Availability',
       'Runtime', 'Director', 'Writer', 'View Rating', 'Release Date Day',
       'Release Date Month', 'Release Date Year', 'Netflix Release Date Day',
       'Netflix Release Date Month', 'Netflix Release Date Year'],axis=1,inplace=True)
```

```python
x1.columns
```

```python
#doing min max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x1=scaler.fit_transform(x1)
```

```python
#we have to convert this array into dataframe
x1=pd.DataFrame(x1,columns=['IMDb Score', 'Rotten Tomatoes Score', 'Awards Nominated For',
       'IMDb Votes'])
```

```python
#merging this scaled data
x2=x.drop(['IMDb Score', 'Rotten Tomatoes Score', 'Awards Nominated For',
       'IMDb Votes'],axis=1)
x=pd.concat([x1,x2],axis=1)
x.head()
```

```python
#splitting data for training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

# **DecisionTreeRegressor**

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=42)
model.fit(x_train, y_train)

y_pred3 = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred3)
r2 = r2_score(y_test, y_pred3)
print(f"Mean Squared Error: {mse}")
print(f"R-squared Score: {r2}")
```

# **XGBoost** **model**

```python
import xgboost as xgb
```

```python
simple_reg = xgb.XGBRegressor(
    objective="reg:squarederror",
    seed=123,
    learning_rate=0.1,
)
#objective="reg:squarederror" to minimize the squared error
#Seed=123 ensure that the results will be the same if the code is run again with the same data.
#modelâ€™s speed at learning from mistakes.learning_rate=0.1 it can prevent the model from making big mistakes by overreacting to errors.
simple_reg.fit(x_train,y_train)

y_pred1 = simple_reg.predict(x_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred1))
print(rmse)
```

```python
y_pred1
```

```python
print(r2_score(y_test,y_pred1))
```

```python
train_predictions = simple_reg.predict(x_train)
test_predictions = simple_reg.predict(x_test)
train_r2score = r2_score(y_train, train_predictions)
test_r2score = r2_score(y_test, test_predictions)

print(f"Training r2 score: {train_r2score}")
print(f"Testing r2 score: {test_r2score}")
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_scores = cross_val_score(simple_reg, x, y, cv=5)
print(f"Cross-Validation Scores: {cross_val_scores}")
print(f"Mean Cross-Validation Score: {cross_val_scores.mean()}")
# Cross-validation involves splitting your dataset into several parts and
#then training and validating the model on different combinations of these parts.
```
